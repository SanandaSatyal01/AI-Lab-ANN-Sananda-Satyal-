```python
import random


def generate_data(num_samples=10):
    data = []
    for _ in range(num_samples):
        x1 = random.uniform(0, 1)
        x2 = random.uniform(0, 1)
        x3 = random.uniform(0, 1)
        # Target function y = 2*x1 + 3*x2 - x3 + 5
        y = 2 * x1 + 3 * x2 - x3 + 5
        data.append(([x1, x2, x3], y))
    return data


data = generate_data()
weights = [0.0, 0.0, 0.0]
bias = 0.0
learning_rate = 0.01
max_epochs = 100

def predict(x, weights, bias):
    return sum(x[i] * weights[i] for i in range(len(weights))) + bias

def train(data, weights, bias, learning_rate, max_epochs):
    for epoch in range(max_epochs):
        total_error = 0.0
        for inputs, target in data:
            y_pred = predict(inputs, weights, bias)
            error = target - y_pred
            total_error += error ** 2


            for i in range(len(weights)):
                weights[i] += learning_rate * error * inputs[i]
            bias += learning_rate * error

        mse = total_error / len(data)
        print(f"Epoch {epoch + 1}: Mean Squared Error = {mse:.6f}")

        #  stop early if mse is very low
        if mse < 1e-6:
            break

    return weights, bias

print("Initial weights:", weights, "Bias:", bias)
weights, bias = train(data, weights, bias, learning_rate, max_epochs)
print("Final weights:", weights, "Bias:", bias)
```
