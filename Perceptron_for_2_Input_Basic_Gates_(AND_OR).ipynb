```python


AND =[0,0,0,1]
OR = [0,1,1,1]

inputs=[[0,0],[0,1],[1,0],[1,1]]

user_prompt=input("What to train AND or OR?.Type AND/OR:").strip().lower()
if user_prompt== "and":
    outputs=AND
elif user_prompt=="or":
    outputs=OR
else:
    print("Invalid input")
    exit()
weight=[0]*2
weight[0]=0.6
weight[1]=0.2
threshold=0.5
function=[]
g_list=[]

def print_truth_table():
    print("x1","x2","y",end=' ')
    print()
    for i in range(len(inputs)):
        for j in range(2):
            print(inputs[i][j],end='  ')
        print(outputs[i], end='  ')
        print("")

def predict(inputs, outputs, weight, function:list,):

    global g_list
    g_list.clear()
    function.clear()
    print("Weights are:", weight)
    for i in range(len(inputs)):
        x1,x2=inputs[i]
        function=activation_function(x1,x2,weight,function)
        print(function)
    for i in range(len(outputs)):
       g_p= 0 if function[i] <= threshold else 1
       g_list.append(g_p)
       actual_Y = outputs[i]
       weight_training(g_p,actual_Y,x1,x2)

    print("Table:", g_list)

def weight_training(g_p,actual_Y,x1,x2):
        global weight
        if g_p != actual_Y:

            err=actual_Y-g_p
            delta_weightone=0.1*x1*err
            weight[0]=weight[0]+delta_weightone
            delta_weighttwo=0.1*x2*err
            weight[1]=weight[1]+delta_weighttwo


def activation_function(x1,x2,weight,function):

    g=(float(x1*weight[0])+(x2*weight[1]))
    function.append(g)

    return function

print_truth_table()
print(f"Initial weights are: {weight}")

i=0
while g_list != outputs:
            print(f"Epoch {i}")
            i=i+1
            predict(inputs,outputs,weight,function)

if g_list == outputs:
    print("\nTraining completed successfully!")
else:
    print("\nTraining stopped. Max epochs reached.")

print("Final Weights:", weight)




```
