```python
import numpy as np
import random

n = int(input("Enter number of features (e.g., 4 or 5): "))
learning_rate = 0.01
epochs = 100

# Generate random dataset
np.random.seed(0)
X = np.random.rand(10, n)
true_weights = np.random.uniform(-1, 1, n)
bias = 5

#  y = w1*x1 + w2*x2 + ... + wn*xn + b
y = np.dot(X, true_weights) + bias


weights = np.zeros(n)
b = 0

print("True Weights:", true_weights)
print("True Bias:", bias)

for epoch in range(epochs):
    total_error = 0
    for i in range(len(X)):
        y_pred = np.dot(X[i], weights) + b
        error = y[i] - y_pred
        weights += learning_rate * error * X[i]
        b += learning_rate * error
        total_error += error ** 2
    mse = total_error / len(X)
    print(f"Epoch {epoch + 1}, MSE: {mse:.4f}")

print("\nLearned Weights:", weights)
print("Learned Bias:", b)
```
